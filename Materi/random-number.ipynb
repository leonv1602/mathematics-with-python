# Contoh

\begin{equation}
g(x) := 4+3x
\end{equation}

```python
x = np.linspace(0,5, 10000)
gx = 4+3*x
x_ = np.random.uniform(low = 0, high = 2, size = 10000)
y = np.random.uniform(low=-0, high = 10, size = 10000)
plt.figure(figsize =(16,16))
plt.plot(x,gx, 'g')
plt.xlim(-3,3)
plt.ylim(-25,25)
plt.grid()
plt.scatter(x_,y, 0.5)
plt.show()
```

\begin{equation}
f(x) := \sin(4x) + \cos(\sin(2x^3))
\end{equation}

```python
x = np.linspace(-5,5, 10000)
fx = np.sin(4*x) + np.cos(np.sin(2*x**3))
x_ = np.random.uniform(low = -5, high = 5, size = 10000)
y = np.random.uniform(low=-0.5, high = 2, size = 10000)
plt.figure(figsize =(16,16))
plt.plot(x,fx, 'g')
plt.scatter(x_,y, 0.5)
plt.show()
```
